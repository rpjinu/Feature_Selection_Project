#2.Apply VarianceThreshold:

```python
model=VarianceThreshold(threshold=0.02)
```

```python
# Apply VarianceThreshold to remove low-variance features
df_vt=model.fit(df)
```

```python
#get column pass the variance or not
df_vt.get_support()
```

```python
#get column  name from this df_vt.get_support() values =True store in columns
columns=df.columns[df_vt.get_support()]
columns
```

```python
#transfom the data in variance theshold which is array format
df_vt=df_vt.transform(df)
```

```python
#convert array dataset to   dataframe
df=pd.DataFrame(df_vt,columns=columns)
```

```python
df.shape
```

#3.Correlation feature selection

```python
# Calculate correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of the correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than a threshold (e.g., 0.9)
to_drop_col = [column for column in upper.columns if any(upper[column] > 0.9)]

# Print the columns to be dropped
print("Columns to be dropped:")
print(to_drop_col)
```

```python
df=df.drop(columns=to_drop_col)
```

```python
#check the number of column left
df.shape[1]
```

#4. Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit(df)
```

```python
df_scaled
```

```python
#transform dataset
df_scaled=df_scaled.transform(df)
```

```python
n_components = 50  # Adjust the number of components as needed
pca = PCA(n_components=n_components)
pca_data = pca.fit_transform(df_scaled)
```

```python
# Convert to DataFrame and create new column names
pca_columns = [f'PC{i+1}' for i in range(50)]  # Naming columns as PC1, PC2, ...
pca_df = pd.DataFrame(pca_data, columns=pca_columns)
```
